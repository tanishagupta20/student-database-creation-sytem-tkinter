```python
from tkinter import*
import tkinter as tk 
from tkinter import ttk
import tkinter.messagebox
import os
import json

window=tkinter.Tk()
window.configure(bg="black")
chitkara=tkinter.Label(window,text="Chitkara University",bg="black",fg="white",font=("Arial",25)).place(relx = 0.5, rely = 0.03, anchor = CENTER)

explore=tkinter.Label(window,text="EXPLORE",bg="black",fg="red",font=("Arial",15)).grid(row=0,column=0)
your=tkinter.Label(window,text="YOUR",bg="black",fg="white",font=("Arial",15)).grid(row=1,column=0)
pontential=tkinter.Label(window,text="POTENTIAL",bg="black",fg="red",font=("Arial",15)).grid(row=2,column=0)

cse=tkinter.Label(window,text="Department of Computer Science & Engineering",bg="black",fg="white",font=("Arial",9,"bold")).place(relx=0.38,rely=0.8)

#image

chitkara=tkinter.Label(window,text="STUDENT DATABASE",bg="black",fg="white",font=("Arial",20)).place(relx = 0.5, rely = 0.2, anchor = CENTER)
frame=tk.Frame(window,width=800,height=400)
frame.place(relx = 0.5, rely = 0.49, anchor = CENTER)
tabControl = ttk.Notebook(frame)

tab1 = ttk.Frame(tabControl)
tab2 = ttk.Frame(tabControl)
tab3 = ttk.Frame(tabControl) 
tab4 = ttk.Frame(tabControl) 
tab5 = ttk.Frame(tabControl) 
 

tabControl.add(tab1, text ='New Student') 
tabControl.add(tab2, text ='Display') 
tabControl.add(tab3, text ='Course Creation') 
tabControl.add(tab4, text ='Display Course') 
tabControl.add(tab5, text ='Course Allocation') 
tabControl.pack(expand = 1, fill ="both") 

#tab1

Name=Label(tab1,text="Enter Your Name",font=("Arial",10),justify="left")
Name.grid(row=0,column=0,padx=10,pady=10,sticky="e")
ename=Entry(tab1,width=50)
ename.grid(row=0,column=3)

Rollno=Label(tab1,text="Enter Your Rollno",font=("Arial",10),justify="left")
Rollno.grid(row=1,column=0,padx=10,pady=10,sticky="e")
eroll=Entry(tab1,width=50)
eroll.grid(row=1,column=3)

Gender=Label(tab1,text="Choose Your Gender",font=("Arial",10),justify="left")
Gender.grid(row=2,column=0,padx=10,pady=10,sticky="e")
gen=StringVar()
gen.set("Male")
rmale=Radiobutton(tab1,text="Male",value="Male",variable=gen)
rmale.grid(row=2,column=3,sticky="w")
rfemale=Radiobutton(tab1,text="Female",value="Female",variable=gen)
rfemale.grid(row=2,column=3,sticky="e")

Address=Label(tab1,text="Address for Correspondence",font=("Arial",10),justify="left")
Address.grid(row=3,column=0,padx=10,pady=10,sticky="e")
eaddress=Entry(tab1,width=50)
eaddress.grid(row=3,column=3)

Phone=Label(tab1,text="Phone No",font=("Arial",10),justify="left")
Phone.grid(row=4,column=0,padx=10,pady=10,sticky="e")
ephone=Entry(tab1,width=50)
ephone.grid(row=4,column=3)

Batch=Label(tab1,text="Your Batch",font=("Arial",10),justify="left")
Batch.grid(row=5,column=0,padx=10,pady=10,sticky="e")
cbatch=ttk.Combobox(tab1,state="readonly",width=48)
cbatch["values"]=("Batch 2018","Batch 2019","Batch 2020","Batch 2021","Batch 2022","Batch 2023","Batch 2024")
cbatch.grid(row=5,column=3,sticky="e")

Hostel=Label(tab1,text="Hostel[Y/N]",font=("Arial",10),justify="left")
Hostel.grid(row=6,column=0,padx=10,pady=10,sticky="e")
CheckVar=IntVar()
CheckVar=1
def check():
    global CheckVar
    CheckVar=False if CheckVar==1 else True
chostel=Checkbutton(tab1,text="Click if you need Hostel Facility",variable=CheckVar,command=check,onvalue=1,offvalue=0)
chostel.grid(row=6,column=3,sticky="w")

def save():
    stu={}
    stu1={}
    stu['Students']=list()
    
    name=ename.get()
    rollno=eroll.get()
    gender=gen.get()
    address=eaddress.get()
    phone=ephone.get()
    batch=cbatch.get()
    hostel=False if CheckVar==1 else True 
    
    stu1['name']=name
    stu1['rollno']=rollno
    stu1['gender']=gender
    stu1['address']=address
    stu1['phone']=phone
    stu1['batch']=batch
    stu1['hostel']=hostel
    
    
    if os.path.isfile("C:\\Users\\hp\\student1.json"):
        with open("C:\\Users\\hp\\student1.json",'r')as f:
            stu=json.load(f)
            stu["Students"].append(stu1)
            
        with open("C:\\Users\\hp\\student1.json",'w')as f:
            json.dump(stu,f)
    
    else:
        with open("C:\\Users\\hp\\student1.json",'w')as f:
            stu["Students"].append(stu1)
            json.dump(stu,f)
            
    tkinter.messagebox.showinfo("Save","Your record has been saved")
    

bsave=Button(tab1,text="Save",width=15,command=save)
bsave.grid(row=7,column=1,sticky="e")

def clear():
    ename.delete(0,"end")
    eroll.delete(0,"end")
    eaddress.delete(0,"end")
    ephone.delete(0,"end")
    chostel.deselect()
    cbatch.delete(0,"end")

bclear=Button(tab1,text="Clear",width=15,command=clear)
bclear.grid(row=7,column=2,sticky="w")


#tab2

treev=ttk.Treeview(tab2,selectmode="browse")
treev.pack()

treev["columns"]=("Rollno","Name","Gender","Address","Phoneno","Batch","Hostel")
treev["show"]="headings"
treev.column("Rollno",width=100)
treev.column("Name",width=100)
treev.column("Gender",width=100)
treev.column("Address",width=100)
treev.column("Phoneno",width=100)
treev.column("Batch",width=100)
treev.column("Hostel",width=100)

treev.heading("Rollno",text="Rollno ")
treev.heading("Name",text="Name ")
treev.heading("Gender",text="Gender ")
treev.heading("Address",text="Address ")
treev.heading("Phoneno",text="PhoneNo ")
treev.heading("Batch",text="Batch ")
treev.heading("Hostel",text="Hostel ")

stu={}
if os.path.isfile("C:\\Users\\hp\\student1.json"):
    with open("C:\\Users\\hp\\student1.json") as f:
        stu=json.load(f)
    for i in stu["Students"]:
        treev.insert("","end",values=(i["rollno"],i["name"],i["gender"],i["address"],i["phone"],i["batch"],i["hostel"]))
else:
    messagebox.showinfo("File Error","File Does not Exist")


#tab3
Course=Label(tab3,text="Course ID",width=50)
Course.grid(row=0,column=0,padx=10,pady=30,sticky="e")
ecourse=Entry(tab3,width=40)
ecourse.grid(row=0,column=1)

Course_name=Label(tab3,text="Course Name",width=50)
Course_name.grid(row=1,column=0,padx=10,sticky="e")
ecourse_name=Entry(tab3,width=40)
ecourse_name.grid(row=1,column=1)

def cour():
    cou={}
    cou1={}
    cou['Courses']=list()
    
    course=ecourse.get()
    coursename=ecourse_name.get()
    
    cou1['Course ID']=course
    cou1['Course Name']=coursename
    
    if os.path.isfile("C:\\Users\\hp\\course1.json"):
        with open("C:\\Users\\hp\\course1.json")as file:
            cou=json.load(file)
            cou["Courses"].append(cou1)
            
        with open("C:\\Users\\hp\\course1.json",'w')as file:
            json.dump(cou,file)
    
    else:
        with open("C:\\Users\\hp\\course1.json",'w')as file:
            cou["Courses"].append(cou1)
            json.dump(cou,file)
            
    tkinter.messagebox.showinfo("Save","Your record has been saved")
    
bsave=Button(tab3,text="Save",width=15,command=cour)
bsave.grid(row=7,column=0,sticky="e" ,padx=30,pady=50)
def delete():
    ecourse.delete(0,"end")
    ecourse_name.delete(0,"end")
    
    
bclear=Button(tab3,text="Clear",width=15,command=delete)
bclear.grid(row=7,column=1,sticky="e")

#tab4

treev=ttk.Treeview(tab4,selectmode="browse")
treev.pack()

treev["columns"]=("Course ID","Course Name")
treev["show"]="headings"
treev.column("Course ID",width=200)
treev.column("Course Name",width=200)

treev.heading("Course ID",text="Course ID")
treev.heading("Course Name",text="Couse Name")


cou={}
if os.path.isfile("C:\\Users\\hp\\course1.json"):
    with open("C:\\Users\\hp\\course1.json") as f:
        cou=json.load(f)
    for i in cou["Courses"]:
        treev.insert("","end",values=(i["Course ID"],i["Course Name"]))
else:
    messagebox.showinfo("File Error","File Does not Exist")

#tab5



sroll=Label(tab5,text="Student Rollno",width=50)
sroll.grid(row=0,column=0,padx=10,pady=30,sticky="e")
esroll=Entry(tab5,width=40)
esroll.grid(row=0,column=1)

batch=Label(tab5,text="Your Batch",width=50)
batch.grid(row=1,column=0,padx=10,pady=10,sticky="e")
cbatch=ttk.Combobox(tab5,state="readonly",width=36)
cbatch["values"]=("Programming in C","Problem Solving With Python","Object Oriented Programming in Java","Introduction to Linux","Database Management System","Computer Networks","Advanced Web Technologies")
cbatch.grid(row=1,column=1)

ballocate=Button(tab5,text="Allocate",width=15)
ballocate.grid(row=2,column=0,sticky="e",padx=50,pady=50)

def my_clear():
    esroll.delete(0,"end")
    
b_clear=Button(tab5,text="Clear",width=15,command=my_clear)
b_clear.grid(row=2,column=1,sticky="e",padx=50,pady=50)

window.mainloop()
```
